```python
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('test_images/0 (239).jpg', 0)
print(img.shape)
height, width=img.shape
plt.imshow(img,cmap='gray')
```

```python
imgFlatten = img.reshape(height*width,1)
print(imgFlatten.shape)
```

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters=6)
model.fit(imgFlatten)
labels = model.labels_
print(labels.shape)
label2D=labels.reshape(height,width) 
```

```python
print(img[200:210,200:210])
print('--------------------------------------------')
print(label2D[200:210,200:210])
```

```python
plt.imshow(label2D)
plt.colorbar()
plt.show()
```

```python
import joblib

joblib.dump(model,'brain_tumor_KMeans_model.sav')
```
